```r
## Cash Positions(Enter Manually):
###FHSA
fcashu=0 #in USD
fcashc=49  #in CAD
###TFSA
tcashu=12 #in USD
tcashc=0  #in CAD
###RRSP
rcashu=7 #in USD
rcashc=0  #in CAD
###Non-Reg
ncashu=0 #in USD
ncashc=0  #in CAD
ncashi=0  #in INR
```

```r
## FX Rates (Today)
usdcad = getQuote("CAD=X", from = Sys.Date(),to = Sys.Date(),warnings = FALSE,auto.assign = TRUE)$Last
cadinr = getQuote("CADINR=X", from = Sys.Date(),to = Sys.Date(),warnings = FALSE,auto.assign = TRUE)$Last
usdinr = getQuote("INR=X", from = Sys.Date(),to = Sys.Date(),warnings = FALSE,auto.assign = TRUE)$Last
usdcad
cadinr
usdinr
```

```r
## Cash Inflows/Outflows to Accounts:
cfacc_df = as.data.frame(capmkt_txn[capmkt_txn$txn_type %in% c('cashflow') & capmkt_txn$direction %in% c('inflow','outflow') , ]) #Filtering only cashflow transactions
cfacc_df$date = as.Date(cfacc_df$date, format = "%Y-%m-%d")  #Changing date class to Date

str(cfacc_df)

price_usd = 0
price_cad = 0
price_inr = 0
for (i in 1:nrow(cfacc_df)){
    if (cfacc_df$currency[i]=='cad' & cfacc_df$direction[i]=="inflow"){
        price_usd[i] =  getQuote("CADUSD=X", from = cfacc_df$date[i],to = cfacc_df$date[i],warnings = FALSE,auto.assign = TRUE)$Last * cfacc_df$price[i]
        price_cad[i] =  cfacc_df$price[i]
        price_inr[i] =  getQuote("CADINR=X", from = cfacc_df$date[i],to = cfacc_df$date[i],warnings = FALSE,auto.assign = TRUE)$Last * cfacc_df$price[i]
        }else if(cfacc_df$currency[i]=='cad' & cfacc_df$direction[i]=="outflow"){
        price_usd[i] =  -1*getQuote("CADUSD=X", from = cfacc_df$date[i],to = cfacc_df$date[i],warnings = FALSE,auto.assign = TRUE)$Last * cfacc_df$price[i]
        price_cad[i] =  -1*cfacc_df$price[i]
        price_inr[i] =  -1*getQuote("CADINR=X", from = cfacc_df$date[i],to = cfacc_df$date[i],warnings = FALSE,auto.assign = TRUE)$Last * cfacc_df$price[i]
        }else if(cfacc_df$currency[i]=='usd'& cfacc_df$direction[i]=="inflow"){
        price_usd[i] =  cfacc_df$price[i]
        price_cad[i] =  getQuote("CAD=X", from = cfacc_df$date[i],to = cfacc_df$date[i],warnings = FALSE,auto.assign = TRUE)$Last * cfacc_df$price[i]
        price_inr[i] =  getQuote("USDINR=X", from = cfacc_df$date[i],to = cfacc_df$date[i],warnings = FALSE,auto.assign = TRUE)$Last * cfacc_df$price[i]
        }else if(cfacc_df$currency[i]=='usd'& cfacc_df$direction[i]=="outflow"){
        price_usd[i] =  -1*cfacc_df$price[i]
        price_cad[i] =  -1*getQuote("CAD=X", from = cfacc_df$date[i],to = cfacc_df$date[i],warnings = FALSE,auto.assign = TRUE)$Last * cfacc_df$price[i]
        price_inr[i] =  -1*getQuote("USDINR=X", from = cfacc_df$date[i],to = cfacc_df$date[i],warnings = FALSE,auto.assign = TRUE)$Last * cfacc_df$price[i]
        }else if(cfacc_df$currency[i]=='inr'& cfacc_df$direction[i]=="inflow"){
        price_usd[i] =  getQuote("INRUSD=X", from = cfacc_df$date[i],to = cfacc_df$date[i],warnings = FALSE,auto.assign = TRUE)$Last * cfacc_df$price[i]
        price_cad[i] =  getQuote("INRCAD=X", from = cfacc_df$date[i],to = cfacc_df$date[i],warnings = FALSE,auto.assign = TRUE)$Last * cfacc_df$price[i]
        price_inr[i] =  cfacc_df$price[i]
        }else if(cfacc_df$currency[i]=='inr'& cfacc_df$direction[i]=="outflow"){
        price_usd[i] =  -1*getQuote("INRUSD=X", from = cfacc_df$date[i],to = cfacc_df$date[i],warnings = FALSE,auto.assign = TRUE)$Last * cfacc_df$price[i]
        price_cad[i] =  -1*getQuote("INRCAD=X", from = cfacc_df$date[i],to = cfacc_df$date[i],warnings = FALSE,auto.assign = TRUE)$Last * cfacc_df$price[i]
        price_inr[i] =  -1*cfacc_df$price[i]
        }
        #print(price_usd[i])
    #cfacc_df = cbind(cfacc_df,c(price_cad,price_usd,price_inr))
}
cfacc_df = cbind(cfacc_df,c(price_usd))
colnames(cfacc_df)[12] = c('price_usd')
cfacc_df = cbind(cfacc_df,c(price_cad))
colnames(cfacc_df)[13] = c('price_cad')
cfacc_df = cbind(cfacc_df,c(price_inr))
colnames(cfacc_df)[14] = c('price_inr')
#colnames(cfacc_df) = c('price_usd','price_cad','price_inr')
cfacc_df   
cfacc_ts = as.xts(cfacc_df, date_col = date) #Converting dataframe to Time series
#str(cfacc_ts)

```

```r
##PF Holdings(excluding cash):
pftx_df = as.data.frame(capmkt_txn[capmkt_txn$txn_type %in% c('investment','trading') & capmkt_txn$direction %in% c('buy','sell') , ]) #Filtering only market transactions
tkr = unique(pftx_df$tkr)
pf = data.frame()
#tkr
for (i in tkr){
    ntx = subset(pftx_df, tkr==i)
    q=0
    buyq = sum(subset(ntx, direction == 'buy')$quantity)
    sellq = sum(subset(ntx, direction == 'sell')$quantity)
    cur = unique(pftx_df[pftx_df$tkr %in% c(i), ]$currency)
    as = unique(pftx_df[pftx_df$tkr %in% c(i), ]$asset)
    q = buyq - sellq
    #print(i)
    #print(nrow(ntx))
    #print(q)
    pf = rbind(pf,c(i,as,cur,q))
}
colnames(pf) = c('tkr','asset','currency','quantity')
pf$quantity = as.numeric(pf$quantity)
str(pf)
pf
```

```r
##Adding CMP:

for (i in 1:nrow(pf)){
    pf$cmp[i]= getQuote(pf$tkr[i], from = Sys.Date(),to = Sys.Date(),warnings = FALSE,auto.assign = TRUE)$Last
}
#pf
## Adding market values in three currencies
mkt_usd = 0
mkt_cad = 0
mkt_inr = 0
for (i in 1:nrow(pf)){
    if (pf$currency[i]=='cad'){
        mkt_usd[i] =  getQuote("CADUSD=X", from = Sys.Date(),to = Sys.Date(),warnings = FALSE,auto.assign = TRUE)$Last * pf$quantity[i] * pf$cmp[i]
        mkt_cad[i] =  pf$quantity[i] * pf$cmp[i]
        mkt_inr[i] =  getQuote("CADINR=X", from = Sys.Date(),to = Sys.Date(),warnings = FALSE,auto.assign = TRUE)$Last * pf$quantity[i] * pf$cmp[i]
        }else if(pf$currency[i]=='usd'){
        mkt_usd[i] =  pf$quantity[i] * pf$cmp[i]
        mkt_cad[i] =  getQuote("CAD=X", from = Sys.Date(),to = Sys.Date(),warnings = FALSE,auto.assign = TRUE)$Last * pf$quantity[i] * pf$cmp[i]
        mkt_inr[i] =  getQuote("USDINR=X", from = Sys.Date(),to = Sys.Date(),warnings = FALSE,auto.assign = TRUE)$Last * pf$quantity[i] * pf$cmp[i]
        }else if(pf$currency[i]=='inr'){
        mkt_usd[i] =  getQuote("INRUSD=X", from = Sys.Date(),to = Sys.Date(),warnings = FALSE,auto.assign = TRUE)$Last * pf$quantity[i] * pf$cmp[i]
        mkt_cad[i] =  getQuote("INRCAD=X", from = Sys.Date(),to = Sys.Date(),warnings = FALSE,auto.assign = TRUE)$Last * pf$quantity[i] * pf$cmp[i]
        mkt_inr[i] =  pf$quantity[i] * pf$cmp[i]
        }
        #print(mkt_usd[i])
    #pf = cbind(pf,c(mkt_cad,mkt_usd,mkt_inr))
}
pf = cbind(pf,c(mkt_usd))
colnames(pf)[6] = c('mkt_usd')
pf = cbind(pf,c(mkt_cad))
colnames(pf)[7] = c('mkt_cad')
pf = cbind(pf,c(mkt_inr))
colnames(pf)[8] = c('mkt_inr')
pf
```

```r
# Todays Position(PF + Cash):

## pfmkt = mktval + cash

pfmkt_usd = sum(pf$mkt_usd) + fcashu + (fcashc/usdcad) + tcashu + (tcashc/usdcad) + rcashu + (rcashc/usdcad) + ncashu + (ncashc/usdcad) + (ncashi/usdinr)
mktval_usd = sum(pf$mkt_usd)
cash_usd = pfmkt_usd - mktval_usd

pfmkt_cad = sum(pf$mkt_cad) + (fcashu*usdcad) + fcashc + (tcashu*usdcad) + (tcashc) + (rcashu*usdcad) + (rcashc) + (ncashu*usdcad) + ncashc + (ncashi/cadinr)
mktval_cad = sum(pf$mkt_cad)
cash_cad = pfmkt_cad - mktval_cad

pfmkt_inr = sum(pf$mkt_inr) + (fcashu*usdinr) + (fcashc*cadinr) + (tcashu*usdinr) + (tcashc*cadinr) + (rcashu*usdinr) + (rcashc*cadinr) + (ncashu*usdinr) + (ncashc*cadinr) + (ncashi)
mktval_inr = sum(pf$mkt_inr)
cash_inr = pfmkt_inr - mktval_inr
```

```r
# Calculating Time weighted return as of today:

cfacc_final = cfacc_df[,c('date','price_usd','price_cad','price_inr')]
cfacc_final$price_cad = cfacc_final$price_cad * -1
cfacc_final$price_usd = cfacc_final$price_usd * -1
cfacc_final$price_inr = cfacc_final$price_inr * -1
cfacc_final = rbind(cfacc_final,c(Sys.Date(),pfmkt_usd,pfmkt_cad,pfmkt_inr))
sum(cfacc_final$price_cad)
twr_usd = 100*xirr(cf = cfacc_final$price_usd, d = cfacc_final$date)
twr_cad = 100*xirr(cf = cfacc_final$price_cad, d = cfacc_final$date)
twr_inr = 100*xirr(cf = cfacc_final$price_inr, d = cfacc_final$date)
print(paste("The Time Weighted Return in CAD : ",round(twr_cad,2),"%   ", Sys.Date()))
print(paste("The Time Weighted Return in USD : ",round(twr_usd,2),"%   ", Sys.Date()))
print(paste("The Time Weighted Return in INR : ",round(twr_inr,2),"%   ", Sys.Date()))
```

```r
#TWR Log file:

smry_df_usd = data.frame()
smry_df_usd = rbind(smry_df_usd, c(twr_usd,mktval_usd,cash_usd))
smry_df_usd = cbind(smry_df_usd, Sys.Date())
smry_df_usd = smry_df_usd[,c(4,1,2,3)]
colnames(smry_df_usd)[1] = c('date')

colnames(smry_df_usd)[2] = c('twr_usd')

colnames(smry_df_usd)[3] = c('mktval_usd')
colnames(smry_df_usd)[4] = c('cash_usd')
#colnames(smry_df_usd)[6] = c('cash_usd')
#smry_df_usd
smry_df_usd_ts = as.xts(smry_df_usd, date_col = date)
smry_df_usd_ts
```
